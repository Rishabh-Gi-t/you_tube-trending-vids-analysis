```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

df = pd.read_csv("USvideos.csv")

# Define plot colors
PLOT_COLORS = ["#268bd2", "#0052CC", "#FF52CC", "#b58900", "#003f5c"]

# Set display format for floating point numbers
pd.options.display.float_format = '{:.2f}'.format

# Set seaborn style
sns.set(style="ticks")

# Set matplotlib configurations
plt.rc('figure', figsize=(8, 5), dpi=100)
plt.rc('axes', labelpad=20, facecolor="#ffffff", linewidth=0.4, grid=True, labelsize=14)
plt.rc('patch', linewidth=0)
plt.rc('xtick.major', width=0.2)
plt.rc('ytick.major', width=0.2)
plt.rc('grid', color='#9E9E9E', linewidth=0.4)
plt.rc('font', family='Arial', weight='400', size=10)
plt.rc('text', color='#282828')
plt.rc('savefig', pad_inches=0.3, dpi=300)
```

```python
df["description"] = df["description"].fillna(value="")
```

```python
df.describe()
```

```python
def contains_capitalized_words(s):
    for w in s.split():
        if w.isupper():
            return True
    return False

# Apply the function to create a new column
df["contains_capitalized"] = df["title"].apply(contains_capitalized_words)

# Fix typo in the column name
value_counts = df["contains_capitalized"].value_counts().to_dict()

# Create the pie chart
fig, ax = plt.subplots() 
_ = ax.pie([value_counts.get(False, 0), value_counts.get(True, 0)], labels=['No', 'Yes'],
           colors=['#003f5c', '#ffa600'], textprops={'color': '#040204'}, startangle=45)
_ = ax.axis('equal')
_ = ax.set_title('Title contains capitalized Word?')

plt.show() 
```

```python
df["title_length"] = df["title"].apply(lambda x: len(x))

fig, ax = plt.subplots()
_ = sns.histplot(df["title_length"], kde=False, bins=20, color=PLOT_COLORS[4], alpha=1, ax=ax)
_ = ax.set(xlabel="Title Length", ylabel="Number of Videos", xticks=range(0, 110, 10))
plt.show()
```

```python
fig, ax = plt.subplots()

_ = ax.scatter(x=df['views'], y=df['title_length'], color=PLOT_COLORS[2], edgecolors='#000000', linewidths=0.5)

_ = ax.set(xlabel="Views", ylabel="Title Length")

plt.show()
```

```python

numeric_columns = df.select_dtypes(include=['number', 'bool']).columns
df_numeric = df[numeric_columns]

h_labels = [x.replace('_', ' ').title() for x in list(df_numeric.columns.values)]

fig, ax = plt.subplots(figsize=(10, 6))

_ = sns.heatmap(df_numeric.corr(), annot=True, xticklabels=h_labels, yticklabels=h_labels, cmap=sns.cubehelix_palette(as_cmap=True), ax=ax)

plt.show()
```
